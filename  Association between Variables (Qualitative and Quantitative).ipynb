# Montando a tabela para salário segundo a região de procedência.

```python
# pego a frequência absoluta de cada valor
df['Região de Procedência'].value_counts()
```

```python
# pego a coluna de salário, agrupo por região de procedência e tiro a média do salário em cada grupo da região
media_regiao_procedencia = df['Salario (x Sal Min)'].groupby(df['Região de Procedência']).mean()
media_regiao_procedencia
```

```python
# pegar a variância desse agrupamento:

var_procedencia = df['Salario (x Sal Min)'].groupby(df['Região de Procedência']).var(ddof=0)
print(var_procedencia)
```

```python
# montar a tabela de contigência:

# itens da tabela

regiao_procedencia = {'capital':11, 'interior':12, 'outra':13}
print(regiao_procedencia)
media_procedencia = {'capital': media_regiao_procedencia[0], 'interior': media_regiao_procedencia[1], 'outra': media_regiao_procedencia[2]}
print(media_procedencia)
std_procedencia = {'capital': 5.476653, 'interior': 5.296055, 'outra' : 3.145453}
print(std_procedencia)
var_procedencia = {'capital': var_procedencia[0], 'interior': var_procedencia[1], 'outra' : var_procedencia[2]}
print(var_procedencia)
```

```python

# adicionando o N  à tabela:
regiao_procedencia = ['capital', 'interior', 'outra']
N = [11, 12, 13]
media = [11.455455, 11.550000, 10.445385]
std = [5.476653, 5.296055, 3.145453]
var = [27.267025, 25.710850, 9.132809]

df_tabela_2 = pd.DataFrame({
          'Região' : regiao_procedencia,
          'N' : N,
          'Média' : media,
          'Desvio Padrão' : std,
          'Variância' : var


})

print(df_tabela_2)
```

```python
varInterna = 0 # variância interna, a soma com os pesos, da fórmula. Vou iniciar com zero para iterar depois.
# variância total, geral:

var_total = df['Salario (x Sal Min)'].var(ddof=0)

for N, Var in zip(df_tabela_2['N'], df_tabela_2['Variância']):
  varInterna = varInterna + N * Var

# é como se fosse a média ponderada da variância
Var_barra = varInterna / 36

# caĺculo do R2:
# 1 - (var_barra)/var_total
R2 = 1 - (Var_barra/var_total)
# visualização da variância total:
print(var_total)
# visualização da variância interna:
print(varInterna)
# visualização do R2:
print(R2)
# visualização da variância média (var barra):
print(Var_barra)
```

```python
# cálculo do R2:

import statsmodels.api as sm
from statsmodels.formula.api import ols

modelo = ols('Q("Salario (x Sal Min)") ~ C(Q("Região de Procedência"))', data=df).fit()
r_sqaured = modelo.rsquared
print('Coeficente de determinação (R²):', r_sqaured)
```

```python
# CÁLCULO DO QUI_QUADRADO:

from scipy.stats import chi2_contingency

# Criar uma tabela de contingência a partir das colunas
contingency_table = pd.crosstab(df['Salario (x Sal Min)'], df['Região de Procedência'])

# Exibir a tabela de contingência
print(contingency_table)

# Realizar o teste qui-quadrado
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Exibir os resultados
print(f"Qui-quadrado: {chi2}")
print(f"p-valor: {p}")
print(f"Graus de liberdade: {dof}")
print("Tabela esperada:")
print(expected)
```
