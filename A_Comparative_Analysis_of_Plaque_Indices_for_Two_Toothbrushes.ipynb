# analysis by sex

## Count of values ​​for the 'Gender' column

```python
Bacterias['Sexo'].value_counts()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt



# Transform data to long format
Bacterias_long = pd.melt(Bacterias, id_vars=['Sujeito', 'Sexo'],  #id_vars: These are the columns that you want to keep fixed (not transform). These columns serve as identifiers that remain constant
                         value_vars=["Hugger_Antes_Escovacao", "Convencional_Antes_Escovacao",
                                     "Hugger_Depois_Escovacao", "Convencional_Depois_Escovacao"],
                         var_name="Tipo_Escovacao", value_name="Valores")

# Add columns to identify the type of brush and the moment
Bacterias_long['Momento'] = Bacterias_long['Tipo_Escovacao'].apply(lambda x: 'Antes' if 'Antes' in x else 'Depois')
Bacterias_long['Escova'] = Bacterias_long['Tipo_Escovacao'].apply(lambda x: 'Hugger' if 'Hugger' in x else 'Convencional')

# Create the subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 7))

# before brushing both brushes
sns.boxplot(ax=axes[0], data=Bacterias_long[Bacterias_long['Momento'] == 'Antes'], x="Sexo", y="Valores", hue="Escova", palette='dark') # plot the graph
# changing the boxplot color to a dark palette

axes[0].set_title('plaque index before brushing by gender')
axes[0].set_ylabel('plaque index')
axes[0].set_xlabel('Gender')

# after brushing both brushes
sns.boxplot(ax=axes[1], data=Bacterias_long[Bacterias_long['Momento'] == 'Depois'], x="Sexo", y="Valores", hue="Escova", palette='colorblind') # plot the graph
# changing the boxplot color to a colorblind palette

axes[1].set_title('plaque index after brushing by gender')
axes[1].set_ylabel('plaque index')
axes[1].set_xlabel('Gender')

plt.tight_layout() # adjust the spacing between subplots
plt.show() # show
```

# Correlation Matrix

```python
matrix_corr = Bacterias[['Hugger_Antes_Escovacao' , 'Convencional_Antes_Escovacao', 'Hugger_Depois_Escovacao', 'Convencional_Depois_Escovacao']].corr()
print(matrix_corr)

# graphical representation of the correlation matrix

plt.figure(figsize=(8, 6))
# center=0 to center the color palette around 0
sns.heatmap(matrix_corr, annot=True, cmap='coolwarm', center=0) # annot=True - adds the correlation values ​​in the cells; cmap= defines the color palette
plt.title('Correlation Matrix of Bacterial Plaque Indices')
plt.show()
```
