```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Pesquisa Nacional por Amostra de Domicílios.csv - Pesquisa Nacional por Amostra de Domicílios.csv.csv')
df.head()
```

```python
df.info()
```

```python
# trocando os códigos pelos significados:
# acessando a coluna sexo do dataframe, eu vou alterar os valores 0 e 1 para masculino e feminino, respectivamente. O inplace=True serve para especificar que a modificação será feita no próprio dataframe, em vez de retornar uma cópia dele com a modificação.
df['Sexo'].replace({0:'Masculino', 1: 'Feminino'}, inplace=True)
df
```

```python
df['UF'].replace({
    11:'Rondônia',
    12:'Acre',
    13:'Amazonas',
    14:'Roraima',
    15:'Pará',
    16:'Amapá',
    17:'Tocatins',
    21:'Maranhão',
    22:'Piauí',
    23:'Ceará',
    24:'Rio Grande do Norte',
    25:'Paraíba',
    26:'Pernambuco',
    27:'Alagoas',
    28:'Sergipe',
    29:'Bahia',
    31:'Minas Gerais',
    32:'Espírito Santo',
    33:'Rio de Janeiro',
    35:'São Paulo',
    41:'Paraná',
    42:'Santa Catarina',
    43:'Rio Grande do Sul',
    50:'Mato Grosso do Sul',
    51:'Mato Grosso',
    52:'Goiás',
    53:'Distrito Federal'

}, inplace=True)

df
```

```python
anos_de_estudo = {
    1:'Sem instrução e menos de 1 ano',
    2:'1 ano',
    3:'2 anos',
    4:'3 anos',
    5:'4 anos',
    6:'5 anos',
    7:'6 anos',
    8:'7 anos',
    9:'8 anos',
    10:'9 anos',
    11:'10 anos',
    12:'11 anos',
    13:'12 anos',
    14:'13 anos',
    15:'14 anos',
    16:'15 anos ou mais',
    17:'Não determinados',

}

df['Anos de Estudo'].replace(anos_de_estudo, inplace=True)
df

```

```python
cor = {
    0:'Indígena',
    2:'Branca',
    4:'Preta',
    6:'Amarela',
    8:'Parda',
    9:'Sem declaração'
}

df['Cor'].replace(cor, inplace=True)
df
```

```python
# Remover os pontos de separação de milhares e converter para números inteiros
df['Altura'] = df['Altura'].str.replace('.', '').astype(float).astype(pd.Int64Dtype())
print(df)
```

```python
# caĺculo de méddia de renda:

media = df['Renda'].mean()
media
```

```python
# cálculo de mediana de rendda:

mediana = df['Renda'].median()
mediana
```

```python
# cálculo de moda de renda:

moda = df['Renda'].mode()
moda
```

```python
# cálculo da amplitude de rendda:

amplitudde = df['Renda'].max() - df['Renda'].min()
amplitudde
```

```python
# cálculo da variância de renda:

variancia = df['Renda'].var()
variancia
```

```python
# cálculo do desvio paddrão de renda:

desvio_padrao = df['Renda'].std()
desvio_padrao
```

```python
# cálculo ddo coeficiente de variação de rendda:

coef_var = (df['Renda'].std() / df['Renda'].mean()) * 100
coef_var
```

```python
# cálculo da média por grupo

# para isso, uso o grupbyu

df['Renda'].groupby(df['Sexo']).mean()
```

```python
# cálculo da média por mais de um grupo:

df['Renda'].groupby([df['Cor'], df['Sexo']]).mean()
```

```python
df.groupby(['UF'])['Renda'].mean().sort_values()
```

```python
# não conseguimos aplicar a modde diretamente na groupby, portanto temos que fazer de outra forma (vide celula abaixo)
df['Renda'].groupby(df['Cor']).mode()
```

```python
# forma que consigo a moda sem dar erro por causa da combinação de groupby e mode:

# vou trauzir: estou agrupando por cor. Nesse agrupamento, eu estou iuunterssada na coluna rena, ondde eu vou aplicar uma função lambda que calcula a moda
df.groupby('Cor')['Renda'].apply(lambda x: x.mode())
```
