```python
# Import the library
import seaborn as sns

# Create a boxplot. The boxplot receives the following parameters: date (receives the dataframe);
# x: categorical variable (its name is enclosed in quotation marks);
# y: quantitative variable (its name is enclosed in quotation marks) and color:'color_name'
# the box height (boxplot) is defined by the interquartile range(IQR), which is the difference between Q3 and Q1
# THE MIDDLE LINE REPRESENTS THE MEDIAN OF THE SEPAL LENGTHS FOR EACH CLASS.
# Values other than whiskers are considered atypical, outliers
sns.boxplot(data=X, x="Target", y="sepal length", color='navy')
```

```python
# I import the Statsmodels library, which is used for statistical modeling.
import statsmodels.api as sm
# I import the ols (Ordinary Least Squares) function from the Statsmodels formula API, which allows you to create regression models using R-style formulas.
from statsmodels.formula.api import ols
# The code creates regression models for each quantitative variable as a function of the categorical variable Target.

# Model for Sepal Length
# the ols function of the statsmodels.formula.api module allows me to use formulas in the R language style
# this Q is a function that ensures that the column name is interpreted correctly by the formula,
# regardless of whether it contains spaces or special characters.
# The C() function indicates that Target is a categorical variable.
# THE OLS FUNCTION STILL RECEIVES A PARAMETER CALLED DATA, WHICH RECEIVES THE DF
# THE FIT FUNCTION ADJUSTS THE MODEL TO THE DATA
modelo = ols('Q("sepal length") ~ C(Q("Target"))', data=X).fit()

# R2 CALCULATION - It's the number that tells us how well the regression model is explaining the data we have
# the modelo.rsquared extracts the coefficient of determination from the fitted model (R²) which indicates the proportion of the variability of the dependent
# variable explained by the model
# the R² ranges from 0 to 1, where 0 is: the regression model could not explain any of the data; 1: The model was able to explain 100% of the data and between
# 0 and 1: The model was able to explain part of the data.
r_squared = modelo.rsquared

# view R²
print("Coeficiente de determinação (R^2):", r_squared)

#  Regression Model for Sepal width
modelo = ols('Q("sepal width") ~ C(Q("Target"))', data=X).fit()

# R2 CALCULATION for sepal width
r_squared = modelo.rsquared
# view R²
print("Coeficiente de determinação (R^2):", r_squared)

# Regression Model for petal length
modelo = ols('Q("petal length") ~ C(Q("Target"))', data=X).fit()

# R2 CALCULATION for petal length
r_squared = modelo.rsquared

# view R²
print("Coeficiente de determinação (R^2):", r_squared)


#  Regression Model for petal width
modelo = ols('Q("petal width") ~ C(Q("Target"))', data=X).fit()

# R2 CALCULATION for petal width
r_squared = modelo.rsquared

# view R²
print("Coeficiente de determinação (R^2):", r_squared)
```

# As três celulas a seguir são a plotagem de sepal width com todas as outras medidas.

```python
import pandas as pd # import the library
import matplotlib.pyplot as plt # import the library

categorias=['Iris-setosa', 'Iris-versicolor','Iris-virginica'] # List containing the 3 categories of flowers in the dataset
cores = {'Iris-setosa': 'cyan', 'Iris-versicolor': 'magenta', 'Iris-virginica': 'yellow'} # color dictionary, where the keys are the categories of flowers

# Create the scatter plot
for cat in set(X['Target']): # I scroll through each unique category in the labels column
    # And within that loop, I plot the scatter plot.
    # In this loop, I take all the values of the data in question that have the same cat category as the target category, i.e. I get the data to plot when the
    #categories hit
    plt.scatter(X.loc[X['Target'] == cat, 'sepal length'], # When the category of the Target column is the same as the category of the CAT, I plot the sepal length
                X.loc[X['Target'] == cat, 'petal width'], # When the category of the Target column is the same as the category of the CAT category, I plot the PETAL WIDTH
                c=cores[cat], label=cat) # I set the color of the plot based on the current category (cat). I also define the label based on the current category (cat)

# Add labels and legend
plt.xlabel('sepal length') # label of x axis
plt.ylabel('petal width') # label of y axis
plt.title('Gráfico de Dispersão') # Chart Title
plt.legend() # Chart Legend

# Show the plot
plt.show()
```
