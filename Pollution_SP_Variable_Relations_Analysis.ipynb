```python
df['CO'] = df['CO'].str.replace(',','.').astype(float) # replace commas with periods. convert column to float
```

```python
df['O3'] = df['O3'].str.replace(',','.').astype(float) # replace commas with periods. convert column to float
```

```python
df['temp'] = df['temp'].str.replace(',', '.').astype(float) # replace commas with periods. convert column to float
```

```python
df['umid'] = df['umid'].str.replace(',', '.').astype(float) # replace commas with periods. convert column to float
```

```python
df.info() # CONCISE SUMMARY OF THE BASIS,After Replace and Convert
```

# Data Exploration:


```python
import matplotlib.pyplot as plt # I import the library for data visualization

```

```python
# sactter plot:

plt.scatter(df['CO'], df['O3'], c= '#000000', marker='*')
plt.xlabel('Monóxido de Carbono')
plt.ylabel('Ozônio')
plt.title('CO VS O3') # title
plt.show() # show
```

```python
# scatter plot

plt.scatter(df['CO'], df['temp'], c='#00BFFF', marker='h')
plt.xlabel('Monóxido de Carbono')
plt.ylabel('Temperatura')
plt.title('CO VS Temperatura') # title
plt.show() # show
```

```python
plt.scatter(df['CO'], df['umid'],c = '#00FF00', marker='H')
plt.xlabel('Monóxido de Carbono')
plt.ylabel('Umidade')
plt.title('CO VS Umidade') # title
plt.show() # show
```

```python
plt.scatter(df['O3'], df['temp'], c='#696969', marker='>')
plt.xlabel('Ozônio')
plt.ylabel('Temperatura')
plt.title(  'O3 VS Temperatura') # title
plt.show() # show
```

```python
plt.scatter(df['O3'], df['umid'], c='#6B8E23', marker='<')
plt.xlabel('Ozônio')
plt.ylabel( 'Umidade')
plt.title(  'O3 VS Umidade') # title
plt.show() # show
```

```python
plt.scatter(df['temp'], df['umid'], c='#778899', marker='^')
plt.xlabel( 'Temperatura')
plt.ylabel( 'Umidade')
plt.title(  'Temperatura VS Umidade') # title
plt.show() # show
```

```python
import plotly.express as px # I import the library for data visualization
```

```python
colors = px.colors.qualitative.swatches()
colors.show()
```

```python
fig = px.bar(df,x='CO',y='O3',labels={
    'CO': 'Monóxido de Carbono',
    'O3': 'Ozônio'
},color_discrete_sequence=px.colors.qualitative.Plotly, template='ggplot2', title='Monóxido de Carbono vs Ozônio')
fig.show()
```

```python
fig=px.bar(df, x='CO', y='temp', labels={
    'CO' : 'MONÓXIDO DE CARBONO',
    'temp' : 'TEMPERATURA'
}, color_discrete_sequence=px.colors.qualitative.D3, template='seaborn', title='Monóxido de Carbono vs Temperatura')
fig.show()
```

```python
fig=px.bar(df, x='CO', y='umid', labels={
    'CO' : 'Monóxido de Carbono',
    'umid' : 'Umidade'
},
           color_discrete_sequence=px.colors.qualitative.G10, template='simple_white', title= 'Monóxido de Carbono vs Umidade')
fig.show()
```

```python
fig=px.bar(df, x='O3', y='umid', labels={
    'O3' : 'Ozônio',
    'umid' : 'Umidade'
},
           color_discrete_sequence=px.colors.qualitative.Alphabet, template='plotly_dark', title= 'Ozônio vs Umidade')
fig.show()
```
