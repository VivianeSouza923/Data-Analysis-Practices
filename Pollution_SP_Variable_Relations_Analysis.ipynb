```python
df['CO'] = df['CO'].str.replace(',','.').astype(float) # replace commas with periods. convert column to float
```

```python
df['O3'] = df['O3'].str.replace(',','.').astype(float) # replace commas with periods. convert column to float
```

```python
df['temp'] = df['temp'].str.replace(',', '.').astype(float) # replace commas with periods. convert column to float
```

# Data Exploration:


```python
import matplotlib.pyplot as plt # I import the library for data visualization

```

```python
# sactter plot:

plt.scatter(df['CO'], df['O3'], c= '#000000', marker='*')
plt.xlabel('Monóxido de Carbono')
plt.ylabel('Ozônio')
plt.title('CO VS O3') # title
plt.show() # show
```

Não há correlação linear, entre CO e O3, e nem tendência, segundo o scatter plot

```python
# scatter plot

plt.scatter(df['CO'], df['temp'], c='#00BFFF', marker='h')
plt.xlabel('Monóxido de Carbono')
plt.ylabel('Temperatura')
plt.title('CO VS Temperatura') # title
plt.show() # show
```
