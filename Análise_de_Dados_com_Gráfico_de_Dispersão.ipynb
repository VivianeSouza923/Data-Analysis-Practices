```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/TesteOrange2.csv - TesteOrange2.csv.csv')
df.head()
```

```python
# embaralhar as linhas do df, resetando  index.
df = df.sample(frac=1).reset_index(drop=True)
df.head()
```

```python
# plotar um gráfico de dispersão tendo como coordenas (x, y):

plt.scatter(df['x'], df['y'], c='k') # k is black

#definir oo títuloo:

plt.title('Dados')

#efirni o rótulo doo eixoo x:

plt.xlabel('x')

# definir o rótulo o eixo y:

plt.ylabel('y')

# exibir o gráfico
plt.show()
```
