```python
import pandas as pd # I import the library for data manipulation and visualization
import numpy as np # I import the library to perform mathematical and numerical operations
import seaborn as sns # Import the library to view statistical graphs
import matplotlib.pyplot as plt # Import the library to create charts
import plotly.express as px
```

```python
Bacterias= pd.read_excel('/home/vivi/Downloads/bacterias - Planilha1.csv')
Bacterias
```

```python
sns.boxplot(data=Bacterias, y="Hugger_Antes_Escovacao")
```

```python
sns.boxplot(data=Bacterias, y="Convencional_Antes_Escovacao")
```

```python
# Combine os dados dos dois tipos de escovao em um nico DataFrame
Bacterias_long = pd.melt(Bacterias, value_vars=["Hugger_Antes_Escovacao", "Convencional_Antes_Escovacao"], var_name="Tipo_Escovacao", value_name="Valores")

# Crie o grfico
plt.figure(figsize=(10, 6))
sns.boxplot(data=Bacterias_long, x="Tipo_Escovacao", y="Valores", hue="Tipo_Escovacao")

# Adicione rtulos e ttulo
plt.xlabel("Tipo de Escovao")
plt.ylabel("Valores")
plt.title("Boxplots de Hugger_Antes_Escovacao e Convencional_Antes_Escovacao")

# Mostrar o grfico
plt.show()
```
