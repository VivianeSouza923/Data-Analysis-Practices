```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Pesquisa Nacional por Amostra de Domicílios.csv - Pesquisa Nacional por Amostra de Domicílios.csv.csv')
df.head()
```

```python
df.info()
```

```python
# trocando os códigos pelos significados:
# acessando a coluna sexo do dataframe, eu vou alterar os valores 0 e 1 para masculino e feminino, respectivamente. O inplace=True serve para especificar que a modificação será feita no próprio dataframe, em vez de retornar uma cópia dele com a modificação.
df['Sexo'].replace({0:'Masculino', 1: 'Feminino'}, inplace=True)
df
```

```python
df['UF'].replace({
    11:'Rondônia',
    12:'Acre',
    13:'Amazonas',
    14:'Roraima',
    15:'Pará',
    16:'Amapá',
    17:'Tocatins',
    21:'Maranhão',
    22:'Piauí',
    23:'Ceará',
    24:'Rio Grande do Norte',
    25:'Paraíba',
    26:'Pernambuco',
    27:'Alagoas',
    28:'Sergipe',
    29:'Bahia',
    31:'Minas Gerais',
    32:'Espírito Santo',
    33:'Rio de Janeiro',
    35:'São Paulo',
    41:'Paraná',
    42:'Santa Catarina',
    43:'Rio Grande do Sul',
    50:'Mato Grosso do Sul',
    51:'Mato Grosso',
    52:'Goiás',
    53:'Distrito Federal'

}, inplace=True)

df
```

```python
anos_de_estudo = {
    1:'Sem instrução e menos de 1 ano',
    2:'1 ano',
    3:'2 anos',
    4:'3 anos',
    5:'4 anos',
    6:'5 anos',
    7:'6 anos',
    8:'7 anos',
    9:'8 anos',
    10:'9 anos',
    11:'10 anos',
    12:'11 anos',
    13:'12 anos',
    14:'13 anos',
    15:'14 anos',
    16:'15 anos ou mais',
    17:'Não determinados',

}

df['Anos de Estudo'].replace(anos_de_estudo, inplace=True)
df

```

```python
cor = {
    0:'Indígena',
    2:'Branca',
    4:'Preta',
    6:'Amarela',
    8:'Parda',
    9:'Sem declaração'
}

df['Cor'].replace(cor, inplace=True)
df
```

```python
# cálculo do escore z de renda:

# como calculamos o escore z? z = x - xm / s, ou seja, escore z é igual a x - xmédia dividido por desvio padrão

escore_z_renda = (df['Renda'] - df['Renda'].mean()) / df['Renda'].std()
escore_z_renda
```

```python
df.info()
```

```python
# escore z para idade:

escore_z_idade = (df['Idade'] - df['Idade'].mean()) - df['Idade'].std()
escore_z_idade
```

```python
# escore z para altura:

escoore_z_altura = (df['Altura'] - df['Altura'].mean()) / df['Altura'].std()
escoore_z_altura
```

```python
# adicionar os escores z como coluna na base:

df['Renda Padronizada'] = escore_z_renda
df['Idade Padronizada'] = escore_z_idade
df['Altura Padronizada'] = escoore_z_altura
df
```

```python
# quartis na base:
# variável renda:

# Verificar se há valores ausentes na coluna 'Renda'
print("Valores ausentes na coluna 'Renda':", df['Renda'].isnull().sum())


quartil_1 = np.nanquantile(df['Renda'], 0.25)
quartil_2 = np.nanquantile(df['Renda'], 0.50)
quartil_3 = np.nanquantile(df['Renda'], 0.75)

#visualizar quaris:
print(quartil_1, quartil_2, quartil_3)
```

```python
# renda da média:

media_renda = df['Renda'].mean()
media_renda
```

# **MEDIDA DE ASSIMETRIA**

# primeiro coeficiente de Pearson:

### AS = (MEDIA - MODA) / DEVIO PADRÃO

```python
# PRIMEIRO COEFICIENTE DE PEARSON DA RENDA:

AS_RENDA = (media_renda - df['Renda'].mode()) / df['Renda'].std()
AS_RENDA
```

```python
# primeiro coeficiente de pearson da idade:

AS_idade = (df['Idade'].mean() - df['Idade'].mode()) / df['Idade'].std()
AS_idade
```

```python
# primeiro coeficiente de pearson da altura

AS_altura = (df['Altura'].mean() - df['Altura'].mode()) / df['Altura'].std()
AS_altura
```

# **MEDIDA DE CURTOSE**

## MEDIDA DE CURTOSE: C = (QUARTIL_3 - QUARTIL_1) / 2 * (P90 - P10)

### obs.: esses p90 e p10, são respectivamente, percentil 90 e percentil 10.

```python
# cálculo do percentil 10 de Renda:

print("Valores ausentes na coluna 'Renda':", df['Renda'].isnull().sum())

p10 = np.percentile(df['Renda'], 0.10)
p90 = np.percentile(df['Renda'], 0.90)

C = (quartil_3 - quartil_1) / 2 * (p90 - p10)
C
```
