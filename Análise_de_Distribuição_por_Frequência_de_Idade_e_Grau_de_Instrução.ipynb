```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random # para gerarmos números aleatórios
```

# Geração de dados para idade:

```python
# criação de um vetor vazio
idade = []
# eu vou preencher o vetor idade com 20 números aleatórios dentro de 20 a 78
for i in range(20):
    idade.append(random.randint(20, 78))

idade
```

# Geração de dados para grau de instrução:

```python
# criação do vetor com as categorias do grau de instrução

categorias_grau_instrucao = ['Ef', 'Em', 'Es']

# vetor vazio do grau de instrução
grau_instrucao = []

# para cada idade, eu vou preencher o vetor grau de instrução com uma escolha aleatória das opções existentes do vetor categorias
for i in range(20):
    grau_instrucao.append(random.choice(categorias_grau_instrucao))

grau_instrucao
```

# Montando a base de dados:

```python
df = pd.DataFrame({'Idade': idade, 'Grau de instrução': grau_instrucao}) # recebe um dicionário
df
```

```python
# tabela de frequência para a variável instrução:

frequencia_instrucao = df['Grau de instrução'].value_counts()
frequencia_instrucao
```

```python
# tabela de frequencia para a variável idade:

frequencia_idade = df['Idade'].value_counts()
frequencia_idade
```

# Plotação do gráfico, visualização dos dados:

```python
from itertools import count
# criação da figura e eixos para o gráfico:

fig, ax = plt.subplots()

# definição das categorias de instrução:

categorias = ['Ef', 'Em', 'Es']
# definir a contagem de frequência para cada categooria:

count = [frequencia_instrucao[1], frequencia_instrucao[2], frequencia_instrucao[0]]

# definir rótulos das barras:

bar_labels = ['Ef', 'Em', 'Es']

# definir cores das barras:

bar_colors = ['cyan', 'green', 'gray']

# criação do gráfico e barras:

ax.bar(categorias, count, label=bar_labels, color=bar_colors)

# definir rótulo do eixo y:

ax.set_ylabel('Frequência - Grau de instrução')

# definir título do gráfico:

ax.set_title('Distribuição por frequência')

# definir legenda do gráfico:

ax.legend(title='Instrução')

# exibir o gráfico:

plt.show()
```
